```python
from sklearn.linear_model import LinearRegression
```

# Que es una regresion lineal?
Una regresión intenta predecir un valor numérico a partir de otros datos.

Ejemplos:

- Predecir ingresos según edad.
- Predecir precio según tamaño.

Aquí usaremos regresión lineal: el tipo más simple

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x=df_clean["edad"], y=df_clean["ingresos"])
plt.xlabel("Edad")
plt.ylabel("Ingresos")
plt.title("Relación entre Edad e Ingresos")
plt.show()
```

```python
#MATRIZ DE CORRELACION
#df_clean.corr()
df_clean[["edad", "ingresos"]].corr()
# Si es 0,0 -> nada
# Si es 0.3 -> debil
# Si es de 0.5 - 0.7 -> moderada
# Si es >0.8 -> fuerte
```

```python
from sklearn.linear_model import LinearRegression

df_clean_model = df_clean.copy()
df_clean_model["ingresos"] = df_clean_model["ingresos"].astype(str).str.replace("$", "", regex=False).str.replace(",", "", regex=False).astype(float)

# Todas las regresiones lineales tienen variables de entrada y de salida
X = df_clean_model[["edad"]]      # variable de entrada
y = df_clean_model["ingresos"]    # lo que queremos predecir
modelo = LinearRegression()
modelo.fit(X, y)
```

Tiene relacion con la formula de la pendiente
Que hace fit? Busca los valores de m, y de b que mejor se ajustan a tus datos
Minimiza el error entre los ingresos reales y los ingresos predichos

```python
edad_nueva =[[30]]
prediccion = modelo.predict(edad_nueva)
prediccion
```

```python
# Podemos evaluar el modelo
from sklearn.metrics import r2_score , mean_squared_error
y_pred = modelo.predict(X) #Son los ingresos que el modelo cree segun la edad
#ingresos = β0 + β1 · edad
#β0 → intercepto
#β1 → coeficiente de edad
r2 = r2_score(y, y_pred)
# 0.3 - o.5 aceptable , pero siempre es mejor tener un modelo  0.6 - 0.8
# o muy bueno > 0.8
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
print(r2)
# Regresion Logistica se suele usar -> Accuracy,Precision , F1 score , Rocali , Roc Auc
```
