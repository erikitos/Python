# Funciones que llaman a otras funciones


```python
def saludar(nombre):
    return f"Hola, {nombre}!"

def despedirse(nombre):
    return f"Adios , {nombre}!"

def mensaje_completo(nombre):
    saludo = saludar(nombre)
    despedida = despedirse(nombre)
    return saludo + " " + despedida

print(mensaje_completo("Ana"))
```

```python
# Opcional
def pedir_sabor():
    sabores = ["vainilla", "chocolate", "fresa"]
    print("Sabores disponibles:", ", ", sabores)
    sabor = input("Elige un sabor: ").strip().lower()
    while sabor not in sabores:
        sabor = input("Error, elige un sabor válido: ").strip().lower()
    return sabor

def pedir_cantidad():
    cantidad = int(input("¿Cuántas bolas quieres? "))
    while cantidad <= 0:
        cantidad = int(input("Error, ingresa un número mayor a 0: "))
    return cantidad

def pedir_helado():
    sabor = pedir_sabor()   # llama a pedir_sabor
    cant = pedir_cantidad() # llama a pedir_cantidad
    print(f"Pedido: {cant} bola(s) de {sabor}")
    return cant, sabor

total_pedidos = int(input("¿Cuántos clientes atenderás? "))
for _ in range(total_pedidos):
    pedir_helado()

```

## Funciones Lamba
Una funcion lamba es una funcion acortada
para escribir una funcion de esta forma:

nombreFuncion = lambda parametros : accion

La diferencia entre las funciones lambda y las funciones  normales es que pueden ser  un poco mas limitadas

```python
sumar = lambda x,y : x + y
print(sumar(3,10))
```

```python
descuentoLambda = lambda desc, precio: ((precio * (1-desc)),(precio * desc))
nuevoPresio = descuentoLambda(0.4,500)
print("El precio a pagar es", nuevoPresio[0], ",el descuento que se hizo fue de : " , nuevoPresio[1])
print(type(nuevoPresio))
```

```python
def promediosYmayor(promedios):
    promedio = 0
    for x in promedios:
        promedio += x
    promedio = promedio / len(promedios)
    return promedio, max(promedios)

promedios = []
continuar = ""

while True:
    if len(promedios) > 0:
        continuar = ""
        while continuar not in ("si", "no"):
            continuar = input("Ingrese una respuesta válida: si/no").lower()
        if continuar == "no":
            break

    while True:
        try:
            calif = int(input("Indique la calificación: "))
            if calif < 0 or calif > 100:
                print("Ingrese un valor válido")
            else:
                promedios.append(calif)
                break
        except ValueError:
            print("Ingrese un valor numérico")

prom, mayor = promediosYmayor(promedios)
print("\nResultados:")
print("Promedio:", prom)
print("Mayor:", mayor)
```

```python
# Calculadora -> suma, resta, producto, división y potencia

def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def producto(a, b):
    return a * b

def division(a, b):
    if b == 0:
        return "Error: no se puede dividir entre 0"
    return a / b

def potencia(a, b):
    return a ** b

def menu():
    print("------------MI CALCULADORA ----------")
    print("1. Suma")
    print("2. Resta")
    print("3. Producto")
    print("4. División")
    print("5. Potencia")
    print("6. Salir")

# Programa principal
while True:
    menu()
    try:
        op = int(input("Seleccione una opción: "))
        if op < 1 or op > 6:
            print("Opción inválida")
            continue
    except ValueError:
        print("Ingrese un número válido")
        continue

    if op == 6:
        print("SALIENDO...")
        break

    try:
        a = float(input("Ingrese el primer número: "))
        b = float(input("Ingrese el segundo número: "))
    except ValueError:
        print("Error: Ingrese un número válido")
        continue

    if op == 1:
        print("Resultado:", suma(a, b))
    elif op == 2:
        print("Resultado:", resta(a, b))
    elif op == 3:
        print("Resultado:", producto(a, b))
    elif op == 4:
        print("Resultado:", division(a, b))
    elif op == 5:
        print("Resultado:", potencia(a, b))
    else:
        print("Opción inválida")
```
