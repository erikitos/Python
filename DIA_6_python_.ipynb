```python
# Getters y Setters
class Cuenta:
    def __init__(self, saldo):
        self.__saldo = saldo

    @property
    def saldo(self):  # Getter
    # Permite leer el valor  de _ saldo usando c.saldo
        return self.__saldo
    @saldo.setter # Setter
    def saldo(self,valor):
      # Se ejecuta cuando hacemos c.saldo = valor
      self._saldo = valor
c= Cuenta(300)
print(c.saldo)
c.saldo = 4000


```

```python
# Clase base que representa un vehículo genérico
class Vehiculo:
    def __init__(self, marca):
        self.marca = marca

    # Método que debe ser implementado por las clases hijas
    def mover(self):
        raise NotImplementedError("Este método DEBE ser implementado.")

# Clase Motor que se usará mediante composición
class Motor:
    def __init__(self, caballos):
        self.caballos = caballos

    # Método que devuelve la potencia del motor en HP
    def potencia(self):
        return f"{self.caballos} HP"

# Clase Coche que hereda de Vehiculo
class Coche(Vehiculo):
    def __init__(self, marca, caballos):
        super().__init__(marca)              # Herencia de Vehiculo
        self.motor = Motor(caballos)         # Composición: Coche TIENE un Motor

    def mover(self):                         # Polimorfismo: implementación específica
        return f"El coche {self.marca} avanza con motor de {self.motor.potencia()}"

# Clase Moto que también hereda de Vehiculo
class Moto(Vehiculo):
    def __init__(self, marca, caballos):
        super().__init__(marca)              # Herencia
        self.motor = Motor(caballos)         # Composición

    def mover(self):                         # Polimorfismo
        return f"La moto {self.marca} acelera con motor de {self.motor.potencia()}"

# Clase Bicicleta que hereda de Vehiculo pero no tiene motor
class Bicicleta(Vehiculo):
    def mover(self):                         # Polimorfismo
        return f"La bicicleta {self.marca} avanza con pedaleo humano"

# Función que prueba una lista de vehículos usando polimorfismo
def probar_vehiculos(lista):
    # Cada objeto en la lista responde al método mover() de forma distinta
    for v in lista:
        print(v.mover())

# Lista de instancias de diferentes tipos de vehículos
vehiculos = [
    Coche("Toyota", 150),
    Moto("Yamaha", 90),
    Bicicleta("Trek")
]

# Se prueba el comportamiento polimórfico de cada vehículo
probar_vehiculos(vehiculos)
```
